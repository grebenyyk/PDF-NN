```python
from diffpy.Structure import loadStructure
from diffpy.srreal.pdfcalculator import DebyePDFCalculator, PDFCalculator
from pyobjcryst import loadCrystal
from matplotlib.pyplot import plot, show
import matplotlib.pyplot as plt
import glob
import numpy as np
import os
```

```python
os.chdir('/Users/dimitrygrebenyuk/Yandex.Disk.localized/Working/PDF/Refinements/PDF-Cluster-Prediction/all_data/cifs/new/')
```

```python
files = glob.glob('*.cif')
#files = ['6_AHAFIZ.cif', '2_DELHIM.cif']
```

```python
i = 0
for f in files:
    try:
        model = loadCrystal(f)
        dpc = DebyePDFCalculator()
        dpc.qmax = 11
        dpc.rmax = 20
        r, g = dpc(model, qmin=0.5)
        datagcalc = np.column_stack([r, g])
        np.savetxt(f.replace('.cif', '.dat'), datagcalc, header='r g')
        plot(r, g)
    except Exception as e:
        print(f"An error occurred while processing file {f}: {str(e)}")
        continue
    print(f, ' structure number', i+1)
    i += 1
```
